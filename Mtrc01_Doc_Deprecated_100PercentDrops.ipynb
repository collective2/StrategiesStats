## Sumarization

Well, my hopes are not fulfilled.

Daily average returns are not normaly distributed. 

As long as the number of days increases (recall we investigate from 7 to 180 days returns) the population returns are dragged more and more to the negative territory. The reason are died systems.

The positive part on the right side of the above chart could arouse some hope... But the reality is, that **all** those positive days (180 days return > 0.25) are from long past history. All that happened in 2005-2009 when just 10 systems existed and performed well.

The results are: 
- We don't have normally distributed data for hypothesis testing. 
- The biggest problem are short lived systems.

# Experiments


We know that we do not have a normally distributed data. So we can't do hypothesis tests like "will be an average return greater then...".

We could do just hypothesis like "will be an average return different from the population average return".

Looking at the charts above, I think it does not make sense. Our population average lies between two left-side "hills" and an answer "a sample average return is different from the population" will not help us on our journey.

But to be fair, lets try something. Visualize the above 180 days returns chart selecting systems by Age. 
Let's select systems older than 90 and 180 days and visualize their returns after 180 days .

```r
sql = "select DateStart, avg(Return180) as avgRet180 from ReturnsDataInIntervalsCleaned 
        where Age > 90
        group by DateStart;"
statsDaily_AgeGt90 = getC2ScoreDbData(sql)

sql = "select DateStart, avg(Return180) as avgRet180 from ReturnsDataInIntervalsCleaned 
        where Age > 180
        group by DateStart;"

statsDaily_AgeGt180 = getC2ScoreDbData(sql)
```

```r
populationDataDaily(statsDaily_AgeGt90,180,"avgRet180")
```

```r
populationDataDaily(statsDaily_AgeGt180,180,"avgRet180")
```

As we can see, there is almost no change on pictures. 

Selection by age does have no effect on 180-days returns statistics.
